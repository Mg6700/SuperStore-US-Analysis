```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_excel("P6-SuperStoreUS-2015.xlsx")
```

```python
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
dataset["Product Base Margin"].fillna(dataset["Product Base Margin"].mean(), inplace=True)
```

```python
dataset["Order Priority"].value_counts()
```

```python
dataset["Order Priority"].unique()
```

```python
dataset["Order Priority"].replace("Critical ", "Critical", inplace=True)
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority", data=dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

```python
dataset["Ship Mode"].value_counts()
```

```python
x = dataset["Ship Mode"].value_counts().index
y = dataset["Ship Mode"].value_counts().values
```

```python
plt.figure(figsize=(6,5))
plt.pie(y, labels=x, startangle=201, autopct="%0.1f%%")
plt.legend(loc= 1)
plt.title("Ship Mode %")
plt.savefig("Ship Mode %.jpg")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode",data=dataset, hue="Product Category")
plt.title("Ship Mode Based on Product Category")
plt.savefig("Ship Mode Based on Product Category.jpg")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x="Customer Segment",data=dataset)
plt.title("Customer Segment")
plt.savefig("Customer Segment.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Product Category",data=dataset[dataset["Product Category"]=="Office Supplies"], hue="Product Sub-Category")
plt.title("Office Supplies Sub-Category")
plt.savefig("Office Supplies Sub-Category.jpg")
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.countplot(x="Product Category",data=dataset[dataset["Product Category"]=="Furniture"], hue="Product Sub-Category")
plt.title("Furniture Sub-Category")
plt.savefig("Furniture Sub-Category.jpg")
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.countplot(x="Product Category",data=dataset[dataset["Product Category"]=="Technology"], hue="Product Sub-Category")
plt.title("Technology Sub-Category")
plt.savefig("Technology Sub-Category.jpg")
plt.show()
```

```python
dataset["Order Month"] = dataset["Order Date"].dt.month
```

```python
dataset["Order Month"].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Order Month",data=dataset)
plt.title("Orders Based on Months")
plt.savefig("Orders Based on Months.jpg")
plt.show()
```

```python
sns.barplot(x="Product Category", y="Profit",data=dataset, estimator="sum")
plt.title("Profit Based on Product Category")
plt.savefig("Profit Based on Product Category.jpg")
plt.show()
```

```python
sns.barplot(x="Product Category", y="Product Base Margin",data=dataset, estimator="sum")
plt.title("Product Base Margin Based on Product Category")
plt.savefig("Product Base Margin Based on Product Category.jpg")
plt.show()
```
